Una combinación lineal es la suma del producto entre un vector y un escalar mas otro vector y otro escalar. Y da como resultado un vector que combina a ambos dos sumados
combinaciónlineal.PNG

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def graficarVectores(vecs,cols,alpha=1):
  plt.figure()
  plt.axvline(x=0,color="grey",zorder=0)
  plt.axhline(y=0,color="grey",zorder=0)

  for i in range(len(vecs)):
    x = np.concatenate([[0,0], vecs[i]])
    plt.quiver([x[0]],
               [x[1]],
               [x[2]],
               [x[3]],
               angles= "xy", scale_units="xy",scale=1,
                color=cols[i],
               alpha = alpha)
```

```python
v1 = np.array([2,5])
v2 = np.array([3,2])
v1v2 = 2*v1+1*v2
```

```python
print(v1v2)
```

```python
graficarVectores([v1,v2,v1v2],["orange","blue","red"])
plt.xlim(-1,8)
plt.ylim(-1,8)
```

```python
for a in range(-10,10):
  for b in range(-10,10):
    plt.scatter(v1[0]*a+v2[0]*b, v1[1]*a+v2[1]*b,
                marker = ".",
                color = "orange")

plt.xlim(-100,100)
plt.ylim(-100,100)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")

plt.show()
```

Aqui estamos viendo varias posibles combinaciones lineales de dos vectores. 
Con los vectores correctos nos permite describir un espacio entero.
