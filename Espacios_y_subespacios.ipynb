```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1,1])
v2 = np.array([-1,-1])
```

Al definir ambos vectores como continuos solo obtenemos una recta como elemento en r2. Obtuvimos solo un subespacio r1

```python
for a in range(-10,10):
  for b in range(-10,10):
    plt.scatter(v1[0]*a+v2[0]*b, v1[1]*a+v2[1]*b,
                marker = ".",
                color = "orange")

plt.xlim(-25,25)
plt.ylim(-25,25)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")

plt.show()
```

```python
v1 = np.array([1,0])
v2 = np.array([2,-3])
```

```python
for a in range(-10,10):
  for b in range(-10,10):
    plt.scatter(v1[0]*a+v2[0]*b, v1[1]*a+v2[1]*b,
                marker = ".",
                color = "orange")

plt.xlim(-25,25)
plt.ylim(-25,25)

plt.axvline(x=0,color="grey")
plt.axhline(y=0,color="grey")

plt.show()
```

Con la combinación lineal de dos elementos como estos nos arroja un espacio vectorial r2 con todas las posibilidades. Veamos como r2 en realidad es un subespacio de r3

```python
from mpl_toolkits.mplot3d import Axes3D

v1 = np.array([1,0,0])
v2 = np.array([2,-3,0]) # El tercer numero es un punto eje Z ya que vamos a trabajar
# fuera del plano, en un grafico 3d

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

for a in range(-10,10):
  for b in range(-10,10):
    ax.scatter(v1[0]*a+v2[0]*b, v1[1]*a+v2[1]*b,
               v1[2]*a + v2[2]*b,
               marker = ".",
               color = "orange")
    
ax.set_xlabel("Eje x") # Pongo etiquetas a los ejes
ax.set_ylabel("Eje y")
ax.set_zlabel("Eje z")

# plt.xlim(-25,25)
# plt.ylim(-25,25)

# plt.axvline(x=0,color="grey")
# plt.axhline(y=0,color="grey")

plt.show()
```

Si estamos trabajando en r3 (3 dimensiones) un hiperplano es r2 como el que hicimos arriba. Si estamos trabajando en r2 un hiperplano es la recta
