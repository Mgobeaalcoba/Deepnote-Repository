Sistemas de ecuaciones lineales:
Sobredeterminados: sin solucion
con una única solución (estos trabajamos hasta acá)
con infinitas soluciones

```python
import numpy as np
import matplotlib.pyplot as plt
```

primero sistema de ecuacion sin solución

$$ y=3x+5 $$
$$ y=-1x+3 $$
$$ y=2x+1 $$

```python
x = np.arange(-6,6)
y_1= 3*x+5
y_2= -1*x+3
y_3= 2*x+1
```

```python
plt.figure() # Hecho 100% por mi solo
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.axvline(x=0,ymin=-20,ymax=20)
plt.axhline(y=0,xmin=-6,xmax=6)
```

```python
plt.figure() 
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3) # Hasta acá es obligatorio

# Definir los limites de cada eje
plt.xlim(-8,8)
plt.ylim(-8,8)

# Dibujar los ejes
plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")

# Orden para mostrar el grafico
plt.show() # Me ofrece un gráfico con el que puedo interactuar al menos en Jupiter Notebook
```

No hay ningun punto de X e Y donde se crucen las tres rectas. Por esto es un sistema sobredeterminado. No tiene solución.** Tenemos mas ecuaciones que variables. **

Vamos con los determinados: solo dos ecuaciones

```python
x = np.arange(-6,6)
y_2= -1*x+3
y_3= 2*x+1
```

```python
plt.figure()
plt.plot(x,y_2) # plot es un mapa de coordenadas cartesianas
plt.plot(x,y_3)

plt.xlim(-6,6)
plt.ylim(-6,6)

plt.axvline(x=0, color="black")
plt.axhline(y=0, color="black")

plt.show()
```

En el ejemplo de arriba el sistema tiene solucion y es única
