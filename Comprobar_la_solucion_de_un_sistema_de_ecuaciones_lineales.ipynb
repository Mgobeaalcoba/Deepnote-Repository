# Python puede usarse para comprobar la solución de un sistema de ecuaciones lineales
**Ecuaciones lineales: Aquellas que forman una linea recta en el plano. Es decir, donde X no está multiplicada por otra variable y tmp está elevada a ninguna potencia**
**Si son un sistema de ecuaciones lineales, debo resolverlas de conjunto reemplazar cada variable en la otra ecuación y despejar el resultado de X para luego hallar Y**

```python
import numpy as np
import matplotlib.pyplot as plt
```

Para graficar una función necesitamos poder evaluarla a los largo de varios puntos. Es decri necesitamos definir un array con los valores que va a recorrer la función. 

$$ y=3x+5 $$
$$ y=2x+3 $$
$$ x > y $$

```python
x = np.arange(-5,5,1) # Con arange armo un array entre dos puntos con un intervalo predefinido como tercer parametro
```

```python
y_1 = 3*x +5
y_2 = 2*x +3
```

Ahora estamos listos para graficar

```python
plt.figure() # Creo la figura de matplotlib.pyplot

plt.plot(x, y_1) # Defino el contenido de los ejes
plt.plot(x, y_2) # Defino el contenido alternativo de los ejes

plt.xlim(-5,5) # Defino los limites de mi eje x
plt.ylim(-5,5) # Defino los limites de eje y

plt.axvline(x=0, color="grey") # Agrego mis coordenadas. Como queremos ver el gráfico para la vertical line
plt.axhline(y=0, color="grey") # Agrego mis coordenadas. Como queremos ver el gráfico para la horizontal line
```

El punto en el que se cruzan nuestro sistema de ecuaciones es el que hace verdadera a esas ecuaciones al mismo tiempo. En el grafico sería X=-2 e Y=-1

Vamos ahora a despejar este sistema de ecuaciones: 

$$ y=3x+5 $$
$$ y=2x+3 $$
$$ x > y $$

$$ 2x+3=3x+5 $$
$$ 3-5=3x-2x $$
$$ -2=x $$

Con el valor de X ahora encuentro Y

$$ y=3(-2)+5 $$
$$ y=-6+5 $$
$$ y=-1 $$

**Efectivamente los valores hallados resolviendo graficamente mi sistema de ecuaciones son los mismos que los halladas despejando las variables.**
