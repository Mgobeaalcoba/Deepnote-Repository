```python
A = np.array([[3,1],[2,1]])
print(A)
```

clase14_2PNG.PNG

```python
b = np.array([[1],[1]])
print(b)
```

Ahora buscamos la inversa de A

```python
inversa_A = np.linalg.inv(A)
print(inversa_A)
```

Nos muestra numeros decimales o float

Si (ver ultima foto):

$$ A*x=b $$

entonces: 

$$ x=b*(A^-1) $$

$$ A^-1 = Inversa A $$

Por lo que:

$$ X=InversaA*b $$

donde el por es igual calcular el producto interno

```python
X = inversa_A.dot(b)
print(X)
```

Comprobemos entonces ahora el sistema de ecuaciones lineales reemplazanod X e Y por los valores que nos arrojó la variable X. Es decir: X=0 e Y=1

```python
ecuacion_1 = 3*0+1
print(ecuacion_1)
```

```python
ecuacion_2 = 2*0+1
print(ecuacion_2)
```

**Efectivamente hemos hallado los valores de contacto entre X e Y para ese sistema de ecuaciones usando la inversa de una matriz.** 
La manera de ver esto es calcular el producto interno de A por X y que nos arroje un vector columna 1,1

```python
print(A.dot(X))
```

Nueva confirmación de la efectivadad de la comprobación.

¿Que sucede si cambiamos los resultados de nuestro sistema de ecuaciones pero mantenemos los mismos coeficientes delante de nuestras variables?
¿Nos sirve la misma matriz inversa? ¿O tenemos que hayar otra? 
Nos sirve la misma. 
Veamoslo para el siguiente sistema de ecuaciones:

$$ 3x+y=3 $$
$$ 2x+y=7 $$

Acá entonces A ya la tengo definida. Para encontrar X solo me cambia b por lo que debo redefinirla. La llamaré b2

```python
b2 = np.array([[3],[7]])
print(b2)
```

Como dijimos mas arriba:

$$ X2 = InversaA*b2 $$

```python
X2 = inversa_A.dot(b2)
print(X2)
```

```python
print(A.dot(X2))
```

**De esta forma se comprueba que mientras conserve los coeficientes de mi sistema de ecuaciones, no es necesario volver a hallar la invertida de la matriz por mas que cambie el resultado. **
